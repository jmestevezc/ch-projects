15.	native-country: País de origen. País de origen de la persona.

#### ¿Cuál es la distribución de ingresos (>50K o <=50K) en función de la edad?

## Algoritmo

Para el dataset que estamos analizando, optaremos por utilizar un modelo de regresión logística. Esto implica aplicar una técnica de aprendizaje automático que se utiliza comúnmente para problemas de clasificación binaria, como el que tenemos aquí (predecir si un individuo gana más de $50,000 al año basado en diversas características del censo). A diferencia de la regresión lineal, que se utiliza para predecir valores continuos, la regresión logística se ajusta a modelos logísticos que calculan la probabilidad de que una observación pertenezca a una clase particular.

```python
# Supongamos que tienes tu conjunto de datos cargado en un DataFrame df

# Dividir el conjunto de datos en características (X) y la variable objetivo (y)
X3 = X2[['age', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week',
        'workclass_Self-emp-not-inc', 'occupation_Exec-managerial',
        'relationship_Not-in-family', 'relationship_Other-relative',
        'relationship_Own-child', 'relationship_Unmarried']]
y2 = df2['income2']  # Suponiendo que 'income' es tu variable objetivo

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X3, y2, test_size=0.2, random_state=42)

# Opcional: Normalizar las características (si es necesario)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Inicializar el modelo de regresión logística
model = LogisticRegression()

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```
